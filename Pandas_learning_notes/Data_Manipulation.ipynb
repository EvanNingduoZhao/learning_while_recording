```python
import pandas as pd
```

```python
pokemon = pd.read_csv("pokemon_data.csv")
pd.options.display.max_rows=7
```

# Sorting

```python
#sort the rows base on two columns,Type 1 is ascending and HP is descending
pokemon.sort_values(['Type 1','HP'],ascending=[1,0])
```

# Making changes to dataframe

```python
# sum up HP	Attack	Defense	Sp. Atk	Sp. Def	Speed columns to creat a total stats column
pokemon['Total']=pokemon['HP']+pokemon['Attack']+pokemon['Defense']+pokemon['Sp. Atk']+pokemon['Sp. Def']+pokemon['Speed']
pokemon.head(5)

```

```python
# drop this added column first
pokemon.drop(columns=['Total'],inplace=True)
#add it back in another way
# iloc("index location") requires rows and cols input, loc uses condition input
pokemon['Total']=pokemon.iloc[:,4:10].sum(axis=1)
pokemon.head(5)
```

```python
# one way to reorder the columns
cols=list(pokemon.columns.values)
pokemon=pokemon[cols[0:4]+[cols[-1]]+cols[4:12]]
pokemon.head(5)
```

# Filtering Data

```python
# if applying two conditions with "and" or "or" relationship
# instead of using the python style just type 'and' or 'or'
# in Pandas we need to use '&' and '|'
# and note, each condition has to be surronded by parenthesis 
grass_poison_pokemon=pokemon.loc[(pokemon["Type 1"]=="Grass")&(pokemon["Type 2"]=="Poison")]

# the above manipulation will filter out the rows you want but keep their old index
# to get 连续的 index you do the following
grass_poison_pokemon.reset_index(drop=True,inplace=True)
grass_poison_pokemon
```

```python
# to filter rows whose one column contains some sepecific words
pokemon.loc[pokemon['Name'].str.contains("Mega")]
```

```python
# and if you want rows whose one column DOESNOT contains some sepecific words
pokemon.loc[~pokemon['Name'].str.contains("Mega")]
```

```python
# use regular expression to do the contains function
import re
# example 1
# flags=re.I means ignore cases
pokemon.loc[pokemon['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]
# example 2
# in regular expression ^ means start with,[a-z]* means followed by zero or more lower case letters
# so this means to filter all the names that starts with 'pi'
pokemon.loc[pokemon['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

# Modifying dataframe with filtering conditions

```python
# for all the rows with Type 1 column as 'Fire', change "Fire" to "Flamer"
pokemon.loc[pokemon['Type 1']=='Fire','Type 1']="Flamer"
pokemon.loc[pokemon['Type 1']=='Flamer','Type 1']="Fire"
# for all the rows with Type 1 column as 'Fire', mark them as Legendary
pokemon.loc[pokemon['Type 1']=='Fire','Legendary']=True
pokemon
```

```python
# you can also modify more than one column by passing in a list 
pokemon.loc[pokemon['Type 1']=='Fire',['Legendary','Generation']]=['Test 1','Test 2']
pokemon
```

# GroupBy in Pandas

```python
# group by Type 1 and take each group's mean
pokemon.groupby('Type 1').mean()
# you can also do .sum() .count() 
```

```python
# groupby and take the mean and sort by defense
pokemon.groupby('Type 1').mean().sort_values('Defense',ascending=False)
```

```python
# if you want to do a really neat count without worrying about null values
# here we are doing a group by two cols
pokemon['count']=1
pokemon.groupby(['Type 1','Type 2']).count()['count']
```
