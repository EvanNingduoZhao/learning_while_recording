```python
# you can use parse_dates to tell panda which column in you data is date 
# and panda will parse all date value to a standardized format of timestamp data type 
weather = pd.read_csv('weather.csv',na_values=['NA'],parse_dates=['day'])
```

```python
weather = weather.set_index('day')
weather
```

```python
# use fillna(0) to replace all NA as 0
weather_replace_NaN_with_0 = weather.fillna(0)
weather_replace_NaN_with_0
```

```python
# like identify null values when reading in csv_data we call also pass a dict to fill na to specify we want to
# replace na with what value in each column
weather_replace_NaN_with_dicts = weather.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
weather_replace_NaN_with_dicts
```

```python
# use method='ffill' means that replace null values with the value of the previous non-null value in the same column
# there is a corresponding 'bfill' mean replace null values with the value of later non-null value 
# these two method is just the same with assign values to null values according to the proportion of non-null values
 
# axis='rows' means we are copying from the same column of the previous row, default is 'rows'
# if you specify as axis='columns' then it means copying from the same row of the previous column
weather_carry_forward = weather.fillna(method='ffill',axis='rows')
weather_carry_forward
```

```python
# apply limit=1 means only the one null closest to the previou non-null get to be replaced
weather_carry_forward_limit_of_1 = weather.fillna(method='ffill',limit=1)
weather_carry_forward_limit_of_1
```

```python
# apply limit=2 means only the one null closest to the previou non-null get to be replaced
weather_carry_forward_limit_of_2 = weather.fillna(method='ffill',limit=2)
weather_carry_forward_limit_of_2
```

```python
weather
```

```python
# interpplate helps to replace null with the average of the newest previous and later non-null values
# this only works for numerical values
weather_interpolate = weather.interpolate()
weather_interpolate
```

```python
weather
```

```python
# if you want to replace null value according to the newest forward and later non-null all considering the gap betwean
# the date of null to each of the two non-null, add method='time', to do this, you have to set date as index
weather_interpolate_time = weather.interpolate(method='time')
weather_interpolate_time
```

```python
# dropna() means as long as a row contains one null value then drop the whole row
weather_if_contain_one_then_drop = weather.dropna()
weather_if_contain_one_then_drop
```

```python
# dropna(how="all") means only if the row all are made of null values except its index, then we drop that row
weather_if_contain_all_then_drop = weather.dropna(how="all")
weather_if_contain_all_then_drop
```

```python
# dropna(thresh=1) means if a row contains 1 or more than 1 non-null value, then we keep it
weather_if_contain_1_then_keep = weather.dropna(thresh=1)
weather_if_contain_1_then_keep
```
