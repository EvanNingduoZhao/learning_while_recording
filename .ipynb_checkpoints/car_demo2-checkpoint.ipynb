```python
df_with34_55.drop(df_with34_55.columns[0],axis=1,inplace=True)
df_with34_55.shape
```

```python
df_with34_55.drop(columns=['CUST_ID',
    'IF_CAR_NETWORK',
                 'CNT_TEL','Role',
                 'dt_Cadillac_CLUB',
                 'Cadillac_point_accu',
                'IF_Buick_CLUB_FLAG',
                'level_Buick',
                'dt_Buick_CLUB',
                'Buick_point_accu',
                'IF_chevy_CLUB_FLAG',
                'level_chevy',
                'dt_chevy_CLUB',
                'chevy_point_accu',
                'CNT_YB_FEE',
                'service_purchase',
                          'IF_CLUB_CAR',
                          'IF_Cadillac_CLUB_FLAG'],axis=1,inplace=True)
df_with34_55.shape
```

```python
df=df_with34_55.drop(columns=['LASTY_CNT_CGBY',
                          'LASTY_AMT_CGBY',
                         'LASTY_CNT_CLDX',
                         'LASTY_AMT_CLDX',
                         'LASTY_CNT_ZBSP',
                         'LASTY_AMT_ZBSP',
                         'LASTY_CNT_NO_ZBSP',
                         'LASTY_AMT_NO_ZBSP',
                         'LASTY_CNT_CGBY_BN',
                         'LASTY_AMT_CGBY_BN',
                         'LASTY_CNT_PTWX_BN',
                         'LASTY_AMT_PTWX_BN',
                             'LASTY_CNT_PTWX_BY',
                             'LASTY_AMT_PTWX_BY',
                             'LASTY_CNT_SGWX',
                             'LASTY_AMT_SGWX',
                             'LASTY_CNT_BUY_YHCP',
                             'LASTY_CNT_NUM_BUY_YHCP',
                             'LASTY_AMT_BUY_YHCP',
                             'LASTY_CNT_WXZ',
                             'LASTY_AMT_JS',
                             'LASTY_CNT_JZ'],axis=1)
df.shape
```

```python
cadillac = df[df['HOLD_GRAND']=='凯迪拉克']
```

```python
cadillac = cadillac[cadillac['IF_ORG_CUST']==0]
```

```python
cadillac.shape
```

```python
cadillac.drop(columns=['Car_model',''],inplace=True)
cadillac.shape
```

```python
cadillac.info()
```

```python
# handling null values
cadillac['Gender']=cadillac['Gender'].fillna(method='ffill')
cadillac['Age']=cadillac['Age'].fillna(method='ffill')
```

```python
#rename the cols with mis-spells
new_cols = cadillac.columns.values
new_cols[1]="first_buy_mon"
cadillac.columns=new_cols
```

```python
# first step construct a list of dictionaries, each dictionary represent a row

# vin number skipped, MSRP waited till packages,if_org is already filtered,HOLD_GRAND all the same
# plan to store used one among the 12 services as a list, with key '12_services' 
# if_type_car all the same,
cadillac_dict = []
for index,row in cadillac.iterrows():
    row_dict = {}
    # vehicles bought less than 18 months are considered new since first maintenance should be prior to that
    if row['first_buy_mon']>0 and row['first_buy_mon']<=18:
        row_dict['first_buy_mon']="new vehicle"
    # between a year and a half and 4 years is consider middle age 
    elif row['first_buy_mon']>18 and row['first_buy_mon']<=48:
        row_dict['first_buy_mon']="middle age vehicle"
    # more than 4 years is considered old since the max in this dataset is 58 and 75% percentile is 37
    else:
        row_dict['first_buy_mon']="old vehicle"
        
    # divide to 13 categories, one is AJX not installed, the other 12 are 12 AJX serveices
    # 12 services can all coexist with the others, AJX not installed can only coexist with CKJCBG（车况检测报告）
    # since a few IF_AJX_cust==0 rows has IF_CKJCBG = 1
    if row['IF_AJX_cust']==0:
        if row['IF_CKJCBG']==0:
            row_dict['IF_AJX_cust']='AJX not installed'
        else:
            row_dict['IF_AJX_cust']='AJX not installed but 车况检测报告'
    
    
    # handling gender
    if row['Gender']=='男':
        row_dict['Gender'] = 'Male'
    else:
        row_dict['Gender'] = 'Female'
        
    
    # 
    if row['Age']<=22 and row['Age']!=-1:
        row_dict['Age'] = 'School Age'
    elif row['Age']>22 and row['Age']<30:
         row_dict['Age'] = 'Age 20s'
    elif row['Age']>=30 and row['Age']<40:
         row_dict['Age'] = 'Age 30s'
    elif row['Age']>=40 and row['Age']<50:
         row_dict['Age'] = 'Age 40s'
    elif row['Age']>=50:
         row_dict['Age'] = 'Age 50 Older'
    
    cadillac_dict.append(row_dict)

```
